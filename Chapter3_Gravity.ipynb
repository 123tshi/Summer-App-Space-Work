````python
# These imports will be used later in the code. Leave them here.
import random
import numpy
import csv
from nasa import earth
import os
from IPython.display import display
from matplotlib import pylab 
# Your code here:

os.environ["NASA_API_KEY"]="F39Ns48L1oD5eCgTLurLmZFipW6UUyq2t5RkAep6"

class GravitationalWaveDetector(object):
    
    def __init__(self, location):
        self.arm_length_km = 4.0
        self.location = location
    
    def gravitational_wave_template(self):
        """ 
        This returns one example of what a gravitational wave should look like. 
        """
        return numpy.genfromtxt('fig1-expected-L.txt')[:,1]

    def get_detector_time(self):
        """
        This returns the time of the data signal collected
        """
        return numpy.genfromtxt('uniform-noise-L.txt')[:,0]
    
    def get_detector_noise(self):
        """
        This returns the detector noise. In reality, LIGO's noise is much more complicated. 
        Here we read in an artificial file where we model it as random noise.
        """
        return numpy.genfromtxt('uniform-noise-L.txt')[:,1]
    
    def get_detector_data(self):
        """ 
        This returns the data signal that the gravitational wave detector is reading.
    
        It is currently set up so that half the time there is a gravitational wave in the data signal.
        In reality gravitational waves are rarely in the data signal.
        """        
        signal_injection = self.gravitational_wave_template()
        zeros = numpy.zeros(len(signal_injection))
        return self.get_detector_noise() + random.choice([zeros,signal_injection])

    def satview(self, startdate, enddate):
       '''
       This displays the first NASA satellite image for the given time period ('YYYY-MM-DD') and location (latitude, longitude).
       '''
       assets = earth.assets(lat=self.location[0], lon=self.location[1], begin=startdate, end=enddate)
       
       if assets:
            display(assets[0].get_asset_image().image)

        
    def plotdata(self, title):
        '''
        This plots the data specific to the location and asks for the title as input.
        '''
        pylab.plot(self.get_detector_time(), self.get_detector_noise(), color="green", label="Noise")
        pylab.plot(self.get_detector_time(), self.get_detector_data(), color="blue", label="Data")
        pylab.plot(self.get_detector_time(), self.gravitational_wave_template(), color="red", label="Expected Data")
        pylab.legend(loc=1)
        pylab.title(title)
        pylab.xlabel("Time")
        pylab.ylabel("Gravitational Wave")
        pylab.show()        
    
    def did_i_detect_gravitational_wave(self):
        '''
        checks if template and detected data are close
        '''
        return numpy.allclose(self.get_detector_data().astype("float64"), self.gravitational_wave_template().astype("float64"), atol=1.5)
        
class VirgoGravitationalWaveDetector(GravitationalWaveDetector):
    
    def __init__(self, location):
        '''Uses superclass GravitationalWaveDetector but speaks italian and has an arm length of 3.0 km.'''
        super().__init__(location)
        self.arm_length_km = 3.0
    
    def speak_italian(self, numbertimes):
        '''returns "ciao mondo" the input number of times'''
        for i in range(numbertimes):
            print('ciao mondo')
    
````
